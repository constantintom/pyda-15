Исходные данные имеют следующую структуру:

1. перечень всех документов

# Задание 1 и Задание 2

````python
### перечень всех документов
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
# перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def main():
    ''' Бесконечный цикл запросов команд'''
    while True:
        print('Введите команду:')
        command = input()
        if command == 'q':
            print('До свидания!')
            break        
        elif command == 'p' or command == 's' or command == 'ad' or command == 'd' or command == 'm':
            doc_doc(documents, directories, command)
        elif command == 'l':
            all_information(documents, directories)
        elif command == 'ads' or command == 'ds':
             self(directories, command)
        

def input_number(n, documents, directories): 
    for i in documents:
            if i['number'] == n:       
                for k, v in directories.items():
                    if i['number'] in v:
                        return i['number'], i['type'], i['name'], k, documents.index(i)
    
                       
def doc_doc(documents, directories, command):
    ''' Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру/
    Пункт 2. Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится'''
    print("Введите номер документа:")
    n = input()
    x = input_number(n, documents, directories)
    if x != None:
        if command == 'p':
            return print('Владелец документа:', x[2])
        elif command == 's':
            return print(f'Документ хранится на полке: {x[3]}')
        elif command == 'd':
            del documents[x[4]]
            directories.setdefault(x[3]).remove(n)
            print('Документ удален. Текущий список документов:')
            all_information(documents, directories)
        elif command == 'm':
            print('Введите полку для хранения:')
            self_number = input_self(input(), directories, command)
            if self_number != None:
                directories.setdefault(self_number).append(n)
                directories.setdefault(x[3]).remove(n)
                print('Документ перемещен. Текущий список документов:')
                all_information(documents, directories)
    elif command == 'p' or command == 's':
        print('Документ не найден в базе')
    elif command == 'ad':
        print('Введите тип документа:')
        types = input()
        print('Введите владельца документа:')
        name = input()
        print('Введите полку для хранения:')
        self_number = input_self(input(), directories, command)
        if self_number != None:
            documents.append({'type': types, 'number': n, 'name': name})
            directories.setdefault(self_number).append(n)
            print('Документ добавлен. Текущий список документов:')
            all_information(documents, directories)
    elif command == 'd' or command == 'm':
        print('Документ не найден в базе. Текущий список документов:')
        all_information(documents, directories)

def all_information(documents, directories):
    '''Пункт 3. Пользователь по команде "l" может увидеть полную информацию по всем документам'''
    for i in documents:
        x = input_number(i['number'], documents, directories)
        print(f'№: {x[0]}, тип: {x[1]}, владелец: {x[2]}, полка хранения: {x[3]}')   

def input_self(s, directories, command):
    if s in directories.keys():
        return s
    elif command == 'ds' or command == 'm':
        print(f'Такой полки не существует. Текущий перечень полок: {", ".join([k for k in list(directories.keys())])}.')
    elif command == 'ad':
        print('Такой полки не существует. Добавьте полку командой ads')
        print('Текущий список документов:')
        all_information(documents, directories)
    elif command == 'ads':
        directories.setdefault(s, [])
        print(f'Полка добавлена. Текущий перечень полок: {", ".join([k for k in list(directories.keys())])}.')   

def  self(directories, command):
    '''Пункт 4. Пользователь по команде "ads" может добавить новую полку'''
    ''' Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)'''
    print('Введите номер полки:')
    self_number = input_self(input(), directories, command)      
    if command == 'ads':
        return print(f'Такая полка уже существует. Текущий перечень полок: {", ".join([k for k in list(directories.keys())])}.')
    if command == 'ds':
        if self_number != None and directories.get(self_number) == []:
            del directories[self_number]
            return print(f'Полка удалена. Текущий перечень полок: {", ".join([k for k in list(directories.keys())])}.')                      
        elif self_number != None:
            return print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: \
{", ".join([k for k in list(directories.keys())])}.') 
    if command == 'ad' or command == 'm':
        return self_number 
main()
````
