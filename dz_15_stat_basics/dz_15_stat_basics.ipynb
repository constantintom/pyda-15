Задание 2. Работа с выбросами
В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

1. В столбце age выявлено 24 выброса со значением = 9, которое является ошибочно введённым значением. И нет ни одной записи со значением 2 (молодая лошадь). Принимаем решение о замене 9 на 2. Отмечаем снижение среднеквадратичного отклонения с 2,17 до 0,27.

```python
horse_data.age.replace(9, 2, inplace = True)
horse_data_old = horse_data.describe()
print(pd.DataFrame(horse_data.age.describe()).transpose())
```

2. Найдем остальные выбросы через межквартильный размах

```python
q1 = horse_data.quantile(0.25)
q3 = horse_data.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
# По ректальной тем-ре
remove_outliers = horse_data[horse_data.rectal_temperature.between(lower_bound.rectal_temperature, upper_bound.rectal_temperature, inclusive = True)]
remove_outliers.describe() - horse_data.describe()
```

```python
remove_outliers.round().mode() - horse_data.round().mode()
# Отмечаем, что значительно изменилась  мода тем-ры конечностей. Удаление выброса может повлиять на оценку. 
```

```python
#  Выбросы по пульсу
remove_outliers = horse_data[horse_data.pulse.between(lower_bound.pulse, upper_bound.pulse, inclusive = True)]
remove_outliers.describe() - horse_data.describe()
```

```python
remove_outliers.round().mode() - horse_data.round().mode()
```

```python
#  Выбросы по Общему белку брюшной полости
remove_outliers = horse_data[horse_data.abdomcentesis_total_protein.between(lower_bound.abdomcentesis_total_protein, upper_bound.abdomcentesis_total_protein, inclusive = True)]
remove_outliers.describe() - horse_data.describe()
```

```python
remove_outliers.round().mode() - horse_data.round().mode()
```

```python
pd.concat([horse_data, remove_outliers]).drop_duplicates(keep=False)
```

### Выводы:
    Применяем замену выбросов к столбцу Возраст.
    Остальные выбросы носят естественный характер, связанный вероятно с болезнями лошадей в острой форме с превышением показателей пульса, ректальной тем-ры.
    Удалять данные выбросы нецелесообразно.
    По столбцу Общий белок брюшной полости кол-во пропусков составляет около 2/3, удалять выбросы нецелесообразно.

Задание 3. Работа с пропусками
Рассчитать количество выбросов для всех выбранных столбцов. Принять и обосновать решение о методе работы с пропусками по каждому столбцу, сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
horse_data.isna().sum()
```

```python
def procent_isna(df):
    for col in df.columns:
        pct_missing = df[col].isnull().mean()
        print(f'{col} - {pct_missing :.1%}')
procent_isna(horse_data)
```

```python
# Проверяем пропуски по Хирургии
horse_data.loc[horse_data.surgery.isnull()].head()
```

```python
horse_data.loc[(horse_data.rectal_temperature == 38) & (horse_data.pain == 4.0)].sort_values('age')
```

Обнаружены пропуски в строке 132 столбец Хирургия и Результат.
Решение: Находим группы записей с одинаковыми параметрами rectal_temperature и боль.
В 2-х записях из 3-х с одинаковым уровнем боли было хиругическое вмешательство - заменим NaN на 1.
В 2-х записях из 3-х лошади жили (включая повышенный пульс по строке 248) - заменим NaN на 1.

```python
horse_data.surgery.fillna(1.0, inplace = True)
horse_data.outcome.fillna(1.0, inplace = True)
horse_data.isna().sum()
```

Пропуски в столбце abdomcentesis_total_protein занимают 66 %, что делает этот столбец не информативным.
Решение: удалить целиком столбец abdomcentesis_total_protein

```python
horse_data.drop(['abdomcentesis_total_protein'], axis = 1, inplace = True)
horse_data.info()
```

По условиям задачи:
с шоком связаны пониженная < 37.8 ректальная тем-ра, повышенный пульс > 40 и пониженная тем-ра конечностей (3, 4);
повышенная тем-ра конечностей  коррелирует с повышенной ректальной температурой
Сгруппируем по этим данным и заменим Ректальную тем-ру (NaN)  и тем-ру конечностей на медиану по группе.

```python
horse_data.rectal_temperature.fillna(horse_data.groupby(['temperature_of_extremities', 'pulse']).rectal_temperature.transform('median'), inplace = True)
horse_data.temperature_of_extremities.fillna(horse_data.groupby(['rectal_temperature']).temperature_of_extremities.transform('median'), inplace = True)
```

Наличие боли связано с необходимостью хирургического вмешательства
Заменим NaN в боли на медиану сгруппировав по хирургическому вмешательству

```python
horse_data.pain.fillna(horse_data.groupby(['surgery']).pain.transform('median'), inplace = True)
horse_data.loc[horse_data.rectal_temperature.isnull()]
```

Заменим NaN в pulse на медиану, сгруппировав по возрасту и боли

```python
horse_data.pulse.fillna(horse_data.groupby(['age', 'pain']).pulse.transform('median'), inplace = True)
horse_data.loc[horse_data.rectal_temperature.isnull()].head()
```

Предположим, что ректальная тем-ра связана с возрастом, наличием боли

```python
horse_data.rectal_temperature.fillna(horse_data.groupby(['age', 'pain']).rectal_temperature.transform('median'), inplace = True)
horse_data.loc[horse_data.temperature_of_extremities.isnull()]
```

Оставшиеся пропуски в температуре конечностей заменим модой

```python
horse_data.temperature_of_extremities.fillna(horse_data.temperature_of_extremities.round().mode()[0], inplace = True)
horse_data.loc[horse_data.temperature_of_extremities.isnull()]
```
