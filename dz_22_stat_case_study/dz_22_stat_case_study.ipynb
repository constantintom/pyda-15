1) Как критики относятся к спортивным играм?  
2) Критикам нравятся больше игры на PC или на PS4?  
3) Критикам больше нравятся стрелялки или стратегии?  

Для каждого вопроса:
- сформулируйте нулевую и альтернативную гипотезы;
- выберите пороговый уровень статистической значимости;
- опишите полученные результаты статистического теста.

````python
import pandas as pd
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
````

````python
data_games = pd.read_csv('D:/DATA_SCIENCE/dataset/22_dz_stat_case_study/video_games_sales.csv')
data_games.head()
````

````python
data_score_sports = data_games[['Genre', 'Critic_Score']][(data_games['Genre'] == 'Sports')]
data_score_sports.fillna(value = float(data_score_sports.Critic_Score.median()), inplace = True)
data_score_sports
````

````python
#  1) Как критики относятся к спортивным играм?
H0: средняя удовлетворенность критиков спортивными играми >= 75%
H1: средняя удовлетворенность критиков спортивными играми составляет менее 75%
````

````python
from scipy import stats as st
alpha = 0.05
result = st.ttest_1samp(data_score_sports['Critic_Score'], 7.5)
print(result)

if (result.statistic < 0) & (result.pvalue / 2 < alpha):
    print('Отвергаем нулевую гипотезу, среднее меньше 7.5')
else:
    print('Не отвергаем нулевую нулевую гипотезу, критики удовлетворены спортивными играми')
````

````python
# 2) Критикам нравятся больше игры на PC или на PS4?
data_score_pc = data_games[['Platform', 'Critic_Score']][(data_games['Platform'] == 'PC')]
data_score_pc.fillna(value = float(data_score_pc.Critic_Score.median()), inplace = True)
data_score_pc
````

````python
data_score_ps4 = data_games[['Platform', 'Critic_Score']][(data_games['Platform'] == 'PS4')]
data_score_ps4.fillna(value = float(data_score_ps4.Critic_Score.median()), inplace = True)
data_score_ps4
````

* H0: средняя оценка критиков игр на PC и PS4 одинаковая
* H1: средние оценки критиков игр на PC и PS4 различаются
* Применяем двухвыборочный двухсторонний критерий.

````python
alpha = 0.05
result = st.ttest_ind(data_score_pc['Critic_Score'], data_score_ps4['Critic_Score'], equal_var=False)
print(result)
if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, критики по разному относятся к играм на PC и PS4')
else:
    print('Не отвергаем нулевую гипотезу')
````

Вывод: критикам больше нравятся игры на PS4 (statistic=-14.27)

````python
#  3) Критикам больше нравятся стрелялки или стратегии?
````

````python
data_score_shooter = data_games[['Genre', 'Critic_Score']][(data_games['Genre'] == 'Shooter')]
data_score_shooter.fillna(value = float(data_score_shooter.Critic_Score.median()), inplace = True)
data_score_shooter
````

````python
data_score_strategy = data_games[['Genre', 'Critic_Score']][(data_games['Genre'] == 'Strategy')]
data_score_strategy.fillna(value = float(data_score_strategy.Critic_Score.median()), inplace = True)
data_score_strategy
````

* H0: средняя оценка критиков игр стрелялок и стратегий одинаковая
* H1: средняя оценка критиков игр стрелялок и стратегий различаются
* Применяем двухвыборочный двухсторонний критерий.

````python
alpha = 0.05
result = st.ttest_ind(data_score_shooter['Critic_Score'], data_score_strategy['Critic_Score'], equal_var=False)
print(result)
if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, критики по разному относятся к стрелялкам и стратегиям')
else:
    print('Не отвергаем нулевую гипотезу')
````

Вывод: критикам больше нравятся стратегии (statistic=-16.52)

````python
data_spam = pd.read_csv('D:/DATA_SCIENCE/dataset/22_dz_stat_case_study/spam.csv')
data_spam.head(8)
````

````python
# 1) Приведите весь текст к нижнему регистру;
data_spam.Message = data_spam.Message.str.lower()
data_spam.head(8)
````

````python
# 2) Удалите мусорные символы;
import re
# Удаляем служебные символы
data_spam.Message = [re.sub('[\W_]+',' ', message) for message in data_spam.Message]
data_spam.head(8)
````

````python
# разделим на слова
data_spam.Message = [message.split() for message in data_spam.Message]
data_spam.head(5)
````

````python
# 3) Удалите стоп-слова;
import nltk
# nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords_set = set(stopwords.words('english'))
data_spam.Message = [[word for word in message if word not in stopwords_set] for message in data_spam.Message]
data_spam.head(5)
````

````python
# 4) Приведите все слова к нормальной форме;
# nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
data_spam.Message = [[wordnet_lemmatizer.lemmatize(word) for word in message] for message in data_spam.Message]
data_spam.head(5)
````

````python
data_spam.Message = [' '.join(message) for message in data_spam.Message]
data_spam.head(5)
````

````python
# 5) Преобразуйте все сообщения в вектора TF-IDF. Вам поможет следующий код:
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(data_spam.Message)
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names())
````

````python
tfidf_matrix
````

````python
# 6) Разделите данные на тестовые и тренировочные в соотношении 30/70, укажите random_state=42. Используйте train_test_split;
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix, data_spam.Category, test_size=0.30, random_state=42)
````

````python
#  7) Постройте модель логистической регрессии, укажите random_state=42, оцените ее точность на тестовых данных;
````

````python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)
````

````python
y_pred = model.predict(X_test)
y_pred
````

````python
# 8) Опишите результаты при помощи confusion_matrix;
````

````python
from sklearn.metrics import confusion_matrix
result = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = result[0][0], result[0][1], result[1][0], result[1][1]
tn, fp, fn, tp
````

````python
print(f' Точность, доля правильных ответов(Accuracy): {(tn+tp)/(tn+tp+fn+fp)}')
print(f' Точность, доля правильно предсказанных ответов (Precision): {tp/(tp+fp)}')
print(f' Полнота, доля результатов предсказанных правильно (Recall): {tp/(tp+fn)}')
print(f' F-мера: {(tp/(tp+fn))*(tp/(tp+fp))*2 / ((tp/(tp+fn)) + (tp/(tp+fp)))}')
````

* Показатель F-меры показывает, что в среднем 82% предсказаний являются верными.
